To perform measures of central tendency Python has helpful packages and tools that can easily calculate the central tendency of variables in a dataset. For each variable, explain, calculate, visualize, and interpret skewness and kurtosis.

Skewness and Kurtosis are both measures that help determine the shape of a probability distribution. Skewness measures the symmetrical or asymmetrical probability distribution of a dataset. Using skewness can help determine if the dataset is more positively or negatively skewed depending on the length of the tail. It is said that if the left tail is longer it is more negatively skewed, and if the tail is longer to the right it is more positively skewed. Kurtosis, similarly measures the tails by how heavy or light they are. Which can be measured by how fat or skinny they are in relation to the normal distribution. 

Interpreting the results of skewness and kurtosis is fairly simple. Skewness determines how symmetrical a variable is, if the skewness results lie within -0.5 to 0.5, it can be deemed a symmetrical dataset. However, if the skewness result is lower or higher, it is deemed asymmetrical. For example, number of bikes has a high skew value indicating that it is not symmetrical. Due to the result being higher than 0.5 it is likely that the dataset has more higher numbers than it does lower numbers. For kurtosis, a higher number can indicate that the distribution is more high, where as a lower number means the distribution is more flat or low. The hour variable has a very small number, meaning that there is not much distribution in the dataset. This is true as the numbers only range from 1-23 as there are only 24 hours in a day. 

```python
from scipy.stats import skew 
from scipy.stats import kurtosis 

print("\n\nSkewness of variables")
print("------------------------------")
print("Number of bikes", skew(rentedBike, axis=0, bias=True))
print("Hour", skew(hour, axis=0, bias=True))
print("Temp", skew(temp, axis=0, bias=True))
print("Humidity", skew(humidity, axis=0, bias=True))
print("Wind", skew(wind, axis=0, bias=True))
print("Visibility", skew(visibility, axis=0, bias=True))
print("Dewpoint", skew(dewPoint, axis=0, bias=True))
print("Solar Radiation", skew(solarRadiation, axis=0, bias=True))
print("Rainfall", skew(rainfall, axis=0, bias=True))
print("Snowfall", skew(snowfall, axis=0, bias=True))

print("\n\nKurtosis of variables")
print("------------------------------")
print("Number of bikes", kurtosis(rentedBike, axis=0, bias=True))
print("Hour", kurtosis(hour, axis=0, bias=True))
print("Temp", kurtosis(temp, axis=0, bias=True))
print("Humidity", kurtosis(humidity, axis=0, bias=True))
print("Wind", kurtosis(wind, axis=0, bias=True))
print("Visibility", kurtosis(visibility, axis=0, bias=True))
print("Dewpoint", kurtosis(dewPoint, axis=0, bias=True))
print("Solar Radiation", kurtosis(solarRadiation, axis=0, bias=True))
print("Rainfall", kurtosis(rainfall, axis=0, bias=True))
print("Snowfall", kurtosis(snowfall, axis=0, bias=True))
```
