The following code will use ANCOVA to help dissertate the differences of patients with BD in comparison to patients with UD or control patients. By providing three different types of ANCOVA we can analyze how covariates differently affect either the group, type, or both. Thus providing crucial evidence as we can determine how well each group has tested with the specified covariate to provide sufficent evidence in our technical report. The following ANCOVA has showed that the p-values for BD were vastly different, indicating that their mean test scores were vastly different from the other patients. Proving the thesis of how patients with BD are more impaired in relation to IPA.


```python
#performing ANCOVA on factor A, 1 categorical 1 continuous 
factorA = ols('Audio_prosody ~ Type', data = miniPons).fit()
print(factorA.summary())
print("\n\n\n\n\n")

#performing ANCOVA on factor B, 2 categorical 1 continuous 
factorB = ols('Face_video ~ Type + Group', data = miniPons).fit()
print(factorA.summary())
print("\n\n\n\n\n")

#performing ANCOVA on factorb C, 2 continuous  1 categorical 
factorC = ols('Right_answers ~ Face_video + Group', data = miniPons).fit()
print(factorC.summary())
```

Calculaing SST, SST used to determine the total variability which serves as a basis to assessing the overall performance of statistical models. SST can also help in evaluating the overall fit and effectiveness of the model. The formula for SST = SSE + SSR.

```python
#Calculating total variability using SST
audioProsody = miniPons['Audio_prosody']
rightAnswer = miniPons['Right_answers']
x = sm.add_constant(audioProsody)

linearModel = sm.OLS(rightAnswer, x).fit()

sse = np.sum((linearModel.fittedvalues - miniPons.Right_answers)**2)

ssr = np.sum((linearModel.fittedvalues - miniPons.Right_answers.mean())**2)

sst = ssr + sse
print("\n\n\n", sst)
```
